### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt
def validate_pie_data(labels, sizes):
    if not isinstance(labels, list) or not isinstance(sizes, list):
        raise ValueError("Both labels and sizes must be lists.")
    if len(labels) != len(sizes):
        raise ValueError("labels and sizes must have the same length.")
    if not all(isinstance(size, (int, float)) for size in sizes):
        raise ValueError("All sizes must be numeric.")
    if sum(sizes) <= 0:
        raise ValueError("The sum of sizes must be greater than zero.")
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
validate_pie_data(labels, sizes)
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title('Pie Chart')
plt.show()
```
