### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt
def validate_scatter_data(x, y):
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("Both x and y must be lists.")
    if len(x) != len(y):
        raise ValueError("x and y must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise ValueError("All elements in x and y must be numeric.")
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
validate_scatter_data(x, y)
plt.scatter(x, y, color='blue', marker='o', s=100, edgecolors='black')
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
