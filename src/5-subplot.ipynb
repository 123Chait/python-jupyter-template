### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
def validate_data(x, y):
    if not isinstance(x, list) or not isinstance(y, list):
        raise ValueError("Both x and y must be lists.")
    if len(x) != len(y):
        raise ValueError("x and y must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise ValueError("All elements in x and y must be numeric.")

def validate_bar_data(categories, values):
    if len(categories) != len(values):
        raise ValueError("categories and values must have the same length.")
def validate_pie_data(labels, sizes):
    if len(labels) != len(sizes):
        raise ValueError("labels and sizes must have the same length.")
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
validate_data(x, y)
validate_bar_data(categories, values)
validate_pie_data(labels, sizes)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b', linewidth=2)
axs[0, 0].set_title('Line Plot')
axs[0, 1].bar(categories, values, color=['cyan', 'magenta', 'yellow', 'purple'], edgecolor='black')
axs[0, 1].set_title('Bar Plot')
axs[1, 0].scatter(x, y, color='g', marker='s', s=100, edgecolors='black')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title('Pie Chart')
plt.tight_layout()
plt.show()
```
